```
#MU
import matplotlib.pyplot as plt
import numpy as np

def espacofinalmu(s0,v,t,t0=0):
  '''Calcula o espaço final para um instante qualquer t em um movimento uniforme
   Uso: (s0,v,t,t0=0)

   Entrada:
   -s0 (float, obrigatório): é o espaço para o instante inicial t0
   -v (float, obrigatório): é a velocidade escalar instantânea e diferente de zero
   -t (float, obrigatório): é o instante final
   -t0 (opcional, padrão: 0): é o instante inicial 

   '''
  s=s0+v*(t-t0)
  return s

def tempomu(s0,s,v,t0=0):
  '''Calcula o tempo para deslocar um determinado espaço em um movimento uniforme
   Uso: (s0,v,t,t0=0)

   Entrada:
   -s0 (float, obrigatório): é o espaço para o instante inicial t0
   -s (float, obrigatório): é o espaço para o instante final
   -v (float, obrigatório): é a velocidade escalar instantânea e diferente de zero
   -t0(opcional, padrão: 0): instante inicial 

   '''
  t=(s-s0)/v+t0
  return t

def velocidademu(s0,s,t,t0=0):
  '''Calcula a velocidade uniforme dado um deslocamento e um tempo
   Uso: (s0,s,t,t0=0)

   Entrada:
   -s0 (float, obrigatório): é o espaço para o instante inicial t0
   -s (float, obrigatório): é o espaço para o instante final
   -t(float, obrigatório): é o instante final
   -t0(opcional, padrão: 0): instante inicial

   '''
  v=(s-s0)/(t-t0)
  return v

def graficomu(s,s0,v,t,Unidade_Espaco,Unidade_Tempo,t0=0):
  '''Cria o gráfico de espaço x tempo em um movimento uniforme
   Uso: (s0,s,t,t0=0)

   Entrada:
    -s (float, obrigatório): é o espaço para o instante final
   -s0 (float, obrigatório): é o espaço para o instante inicial t0
   -s (float, obrigatório): é o espaço para o instante final
    -v (float, obrigatório): é a velocidade escalar instantânea e diferente de zero
   -t(float, obrigatório): é o instante final
   -t0(opcional, padrão: 0): instante inicial

  '''
  xtempo=np.linspace(0,t)

  yespaco=np.array([espacofinalmu(s0,v,t,t0) for t in xtempo])
  plt.figure()
  plt.title('Espaço x Tempo')
  plt.xlabel(f'{Unidade_Tempo}')
  plt.ylabel(f'{Unidade_Espaco}')
  plt.grid()
  plt.plot(xtempo, yespaco)


def mu(escolha,Unidade_Espaco,Unidade_Tempo):

  #espacofinalmu(s0,v,t,t0=0)
  if escolha == 's':
    s0=float(input('Qual o espaço (s0) para o instante inicial t0? '))
    v=float(input('Qual é a velocidade (v)? '))
    t=float(input('Qual o tempo final (t)? '))
    t0=float(input('Qual é instante inicial (t0)? '))

    s=espacofinalmu(s0,v,t,t0)
    print(f'\n O espaço para o instante final é {s} {Unidade_Espaco}\n')
    
    graficomu(s,s0,v,t,Unidade_Espaco,Unidade_Tempo,t0)

  elif escolha =='t':
    #tempomu(s0,s,v,t0=0):
    s0=float(input('Qual o espaço (s0) para o instante inicial t0? '))
    s=float(input('Qual o espaço (s) para o instante final? '))
    v=float(input('Qual a velocidade (v)? '))
    t0=float(input('Qual é instante inicial (t0)? '))

    t=tempomu(s0,s,v,t0)
    print(f'\n O tempo final (t) é {t} {Unidade_Tempo} \n')

    graficomu(s,s0,v,t,Unidade_Espaco,Unidade_Tempo,t0)

  elif escolha =='v':
    #velocidademu(s0,s,t,t0=0)
    s0=float(input('Qual o espaço (s0) para o instante inicial t0? '))
    s=float(input('Qual o espaço (s) para o instante final?'))
    t=float(input('Qual o tempo final (t) ? '))
    t0=float(input('Qual é instante inicial (t0)? '))

    v=velocidademu(s0,s,t,t0)
    print(f'\n A velocidade (v) é é {v} {Unidade_Espaco}/{Unidade_Tempo} \n')

    graficomu(s,s0,v,t,Unidade_Espaco,Unidade_Tempo,t0)
  

#Encontro de dois móveis

def espacoencontromu(sa0,sb0,va,vb,unidadetempo,unidadeespaco,ta0=0,tb0=0):

  '''
  Calcula o espaço e o tempo de encontro de dois móveis que percorrer a mesma trajetória.
   Uso: (sa0,sb0,va,vb,unidadetempo,unidadeespaco,ta0=0,tb0=0)

   Entrada:

   -sa0 (float, obrigatório): é o espaço em que o móvel a está no instante inicial t0
   -sb0 (float, obrigatório): é o espaço em que o móvel b está no instante inicial t0
   -va (float, obrigatório): é a velocidade escalar instantânea do móvel a e diferente de zero
   -vb (float, obrigatório): é a velocidade escalar instantânea do móvel b e diferente de zero
   -ta0(opcional, padrão: 0): instante inicial para o móvel a
   -tb0(opcional, padrão: 0): instante inicial para o móvel b

   -unidadetempo (string, obrigatório): unidade de tempo
   -unidadeespaco (string, obrigatório): unidade de espaço

  '''
  t=(sa0-sb0-va*ta0+vb*tb0)/(vb-va)
  s=sb0+vb*t-vb*tb0
  print(f'\n O tempo para o encontro entre a e b é {t} {unidadetempo} e será na posição {s} {unidadeespaco}\n')

  tempo = np.linspace(0,t+1)
  espacoa= np.array([espacofinalmu(sa0,va,t,ta0) for t in tempo])
  espacob= np.array([espacofinalmu(sb0,vb,t,tb0) for t in tempo])

  plt.figure()
  plt.title('Espaço x Tempo')
  plt.xlabel(f'{unidadetempo}')
  plt.ylabel(f'{unidadeespaco}')
  plt.plot(tempo, espacoa,label='Móvel a')
  plt.plot(tempo, espacob,label='Móvel b')
  plt.grid()
  plt.legend()
```

```
#movimento uniformemente variado
def aceleracao_media(V, V0, T, T0, param,a):
  '''
  Aceleração é uma medida de quanto a velocidade aumenta ou diminui em um dado periodo de tempo
  temos que a aceleração em um dado ponto é: 
  a = dv(t)/dt
  mas ela também pode ser obtida pela variação da velocidade em um dado periodo de tempo, que é o caso dessa função que temos aqui 
  Nessa situação, o calculo é feito entre dois pontos obtidos a partir de sua posição em função do tempo, um ponto inicial e outro final. 

  Os parametros de entrada são:
  Vf : Velocidade final, isto é quando o corpo passa pelo fim do trecho onde se deseja calcular a aceleração
  Vi : Velocidade inicial, isto é quando o corpo passa pelo inicio do trecho onde se deseja calcular a aceleração. 
  Tf : Tempo final, quando o corpo passa pelo ponto inicial, o tempo começa a ser contado, aqui é quando o corpo passa pelo ponto final  
  Ti : Tempo inicial, quando o corpo passa pelo inicial, 

  '''
  G = 0 #variavel de retorno
  if param == 'a':
      A = (V - V0)/(T - T0)
      G = A

  elif param == 'v':
      V = a*(T - T0) + V0
      G = V
  
  elif param == 'vi':
      V0 = V - a*(T-T0)
      G = V0
  
  elif param  == 'ti':
      T0 = T - (V/a) + (V0/a)
      G = T0
  elif param == 'tf':
      T = T0 +(V/a) - (V0/a)
      G = T
  return (G)
#equação horaria da velocidade:
def eq_horaria_vel_muv(V0,a,T,param, T0,V):
    '''
    essa formula serve para calcular a velocidade em um dado tempo, mas tambem pode ser usada para gerar um grafico.
    Ela retorna a velocidade "V" de um corpo em um dado periodo de tempo T. 
    Também serve para calcular a Velocidade Inicial do corpo, sabendo a velocidade final, a aceleração e o tempo T da velocidade final  

    É uma função muito util para nós pois permite fazer 5 calculos diferentes com a mesma função 
    Os parametros de entrada são:
    V : Velocidade
    V0 : Velocidade inicial
    a : aceleração do corpo
    T : Tempo 
    T0 : foi definido como sendo 0 para casos mais comums, mas quando quiser o usuario pode definilo novamente se quiser
    param : serve para dar a escolha do que se quer calcular sendo:
    'v' : para calcular a velocidade V em um dado periodo de tempo. 
    'v0' : para calcular a velocidade inicial do corpo 
    'a' : para calcular a aceleração do corpo sabendo a velocidade final, a iinicial, e o tempo final, quando n se sabe o tempo inicial. 
    't' : para calcular o tempo sabendo v final, v inicial e a aceleração
    't0' : para calcular o tempo inicial onde o movimento do corpo começou a ser levado em conta, ou seja, é o tempo
    '''

    if param == 'v': #calcula a velocidade final
        V = V0 + a*(T - T0)#essa funcao retorna a velocidade do corpo em um dado periodo de tempo T com aceleração a e velocidade inicial V. 
        return (V) 
    elif param == 'v0': #calcula a velocidade inicial do corpo 
        V0 = V - (a*(T-T0))
        return (V0)
    elif param == 'a': #calcula a aceleração do corpo
        a = (V - V0)/(T - T0) 
        return (a)
    elif param == 't': #calcula o tempo que passou para o corpo sair da velocidade inicial para a velocidade final. 
        T =((V - V0) / a ) + T0
        return (T)
    elif param == 'ti':#calcula o tempo incial que foi considerado no movimento do corpo
        T0 = ((V - V0)/a) - T
        return (T0)

# equação do espaço em função do tempo no Movimento Uniformemente Variado:
# S(t) = S0 + V0*T + (1/2)*a*T**2 

def eq_horaria_espaco_muv(S,S0,V0,a,T,T0, param):
    '''
    A equação tem a forma:
    S(t) = S0 + V0*(T - T0) + 1/2*a*(T - T0)**2 (1)
    mas também pode ser escrita assim:
    S(t) = S0 + V0*(T) + 1/2*a*(T)**2 (2)
    
    Na equação (1), temos o termo T0 quando sabemos que T0 != 0, sendo assim, é uma equação completa para calcular:
    - Velocidade Inicial
    - Espaço Final
    - Tempo final
    - Aceleração
    - Espaço inicial
    - Tempo Final 
     
    Os parâmetros dessa função são:
    S0 : espaco inicial
    V0 : Velocidade inicial do corpo
    a : Aceleração do corpo ( não sei se é a aceleração média)
    T : Tempo
    T0 : Tempo Inical 
    param : É um parametro para efetuar um tipo especifico de calculo que essa função é capaz de fazer
    
    os valores possiveis para o parametro são:
    s :calcula o espaço final 
    s0 : Calcula o espaço inicial
    v0 : Calcula a velocidade Inicial   
    a : Calcula a aceleração
    t0 :Calcula o tempo inicial
    t : Calcula o tempo final


    '''
    
    # para calcular o espaco final tendo a, V0, T, T0 e S0:
    if param == 'S': #calcula o espaço para um tempo t 
        S = S0 + V0*(T - T0) + 1/2*a*(T-T0)**2
        return (S)
    elif param == 's0':#calcula o espaço inicial de onde o corpo partiu com relação ao ponto de referencia
        S0 = S - V0*(T - T0) - 1/2*a*(T-T0)**2 
        return (S0)
    elif param == 'v0':#calcula a velocidade que o corpo tinha quando passou pelo espaço inicial
        V0 = (S - S0  - 1/2*a*(T-T0)**2)/(T - T0)
        return (V0)
    elif param == 'a':#calcula a aceleração que o corpo desenvolve durante o seu movimento
        a =(S - S0 - V0*(T - T0))/(1/2*(T-T0)**2)
        return (a)
    elif param == 't0':#seria para calcular o tempo inicial proposto pelo problema tavez, leva muito trabalho fazer isso vai ser dificil. 
          #T0 = S - S0
          #ainda tenho que encontrar uma forma de achar esse t0 acho que vai ser por newton
          T0 = T - (((S-S0) - V0)*2)/a
          return(T0) 
    elif param == 't':
        #same as before
        T = T0 +  (((S-S0) - V0)*2)/a 
        return (T)
    else:
        print('parametro de entrada invalido')#caso algum parametro seja passado de forma incorreta

#funçao da velocidade escalar média:
def vel_escalar_media_muv (V1, V2, param ):
    '''
    A função da velocidade escalar média no Movimento Uniformemente Variado é utilizada para calcular a média da velocidade desenvolvida entre dois pontos
    Sendo eles, um onde a velocidade do corpo é V1 e o segundo onde a velocidade do corpo é V2. 
    sua forma é como se fosse uma média artmética entre dois valores de x:
    Vm = (V1 + V2)/2
    pode ser usada para calcular a velocidade média, a velocidade inicial,e a velocidade final do corpo 

    Os parâmetros são:
    - V1 : Velocidade do corpo no ponto 1 
    - V2 : Velocidade do corpo no ponto 2
    - Vm : Velocidade escalar média desenvolvida pelo corpo
    - param : serve como parametro de escolha do calculo que a função poderá executar ao ser chamada. 

    Os valores de param podem ser :
    - 'vm' : calcula a velocidade média do corpo.
    - 'v1' : calcula a velocidade 1 do corpo.
    - 'v2' : calcula a velocidade 2 do corpo.
    '''    

    if param == 'vm':
        Vm = (V1 + V2)/2
        return (Vm)# calcula a velocidade escalar média do corpo
    elif param == 'v1':
        V1 = Vm*2 - V2 
        return (V1)# calcula a velocidade 1 do corpo
    elif param == 'v2':
        V2 = (Vm*2 - V1)
        return (V2) # calcula a velocidade 2 do corpo


#equação de torriceli: 
def eq_torricelli_muv(V0, DS, a, V, param): #definindo a equação de torriceli 
    ''' 
    Para alguns problemas de cinematica na fisica, a diferença entre resolver um exercicio e não resolve-lo está no ponto de saber ou n o tempo que é discorrido 
    durante o movimento do corpo, ou dos corpos em questão. 
    A equação de torriceli, é muito util quando n se tem uma ideia do tempo que envolve o problema pois ela n depende diretamente dele. apenas da velocidade,
    do espaço e da aceleração. 
    ela tem a seguinte forma:

    (V)² = (V0)² + 2*a*DS;

    Os Parâmetros dessa função são:

    - V : Velocidade final do corpo 
    - V0 : velocidade inicial do corpo
    - a : Aceleração do corpo 
    - DS : Deslocamento do corpo entre V0 e V
    - param : Parametro previamente estabelecido como v para calcular a velocidade 

    '''
    if param == 'v':
        #calcula a velocidade final do corpo 
        V = (V0**2 + 2*a*DS)**(1/2)
        return (V)
    elif param == 'v0':
        #calcula a velocidade inicial do corpo 
        V0 = (V**2 - 2*a*DS)**(1/2)
        return (V0)
    elif param == 'a':
        #calcula a Aceleração do corpo 
        a = (V**2 - V0**2)/2*DS
        return (a)
    elif param == 'DS': 
        #calcula o espaço entre V0 e V 
        DS = ((V**2) - (V0**2))/(2*a)
        return (DS)
    
    
    
#agora temos que definir uma função para calcular o tempo de encontro de dois corpos em ao longo de um trajetoria
        
def encontro_dois_corpos_muv(S01, S02, V01, V02,a1, a2 ):
    '''
    Essa funcao tem o intuito de calcular o tempo de encontro entre dois corpos no muv
    
    esse tempo de encontro é obtido quando a posicao do corpo 1 é a mesma do corpo 2. para isso, vamos fazer:
        
    S1 = S2 #que é quando as posicoes do corpo 1 e o corpo 2 se igualam, ou seja eles se encontram. 
    
    V01*T + (1/2)*a1*(T**2) + S01 = S02 + V02*T + (1/2)*a2*(T**2)
    #passando os termos do segundo corpo para o lado esquerdo, igualando o lado direito a 0, e reagrupando fica: 
    
    V01*T + (1/2)*a1*(T**2) + S01 - (S02 + V02*T + (1/2)*a2*(T**2)) = 0 
    
    T**2((1/2)*a1 - (1/2)*a2) + T * (V01 - V02) + (S01 - S02) = 0 
    
(T**2)*(         a         )  + T * (     b    )  +  (    c    ) = 0 
    
    Se olharmos bem, temos em mãos uma equação do segundo grau. e sabemos como resolve-la, por meio do metodo de bhaskara. 
    
    T1 = -b + (((b**2) - (4 * a * c))**0.5)/(2*a)
    
    T2 = -b - (((b**2) - (4 * a * c))**0.5)/(2*a)
    
    Os parametros da funcao são:
       
       -S01 Posicao inicial do corpo 1
       -S02 Posicao inicial do corpo 2 
       -V01 Velocidade inicial do corpo 1 
       -V02 Velocidade inicial do corpo 2
       -a1 aceleracao do corpo 1
       -a2 aceleracao do corpo 2
       -T tempo
       
    Observacao: como estamos falando de encontro, temos que enfatizar para o usuario que os valores da velocidade e a aceleração
    do corpo que esta contra o sentido da trajetoria do movimento devem ser negativos
    
    '''
    # temos entao que:
    a =((1/2)*a1 - (1/2)*a2)
    b =(V01 - V02)
    c =(S01 - S02)   
    
    
    T1 = -b + (((b**2) - (4 * a * c))**0.5)/(2*a)
    
    T2 = -b - (((b**2) - (4 * a * c))**0.5)/(2*a)
    
    return ( T1 , T2)#T1 e T2 sao as solucoes da equacao 
```
